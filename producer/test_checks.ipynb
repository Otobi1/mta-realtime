```python
import os
import requests
from google.transit import gtfs_realtime_pb2
```

```python
API_KEY = os.getenv("MTA_API_KEY")
base_url = "https://api-endpoint.mta.info/Dataservice/mtagtfsfeeds/"
```

```python
# endpoints = {
#     A/C/E:  "nyct%2Fgtfs-ace" -- works
#     G: "nyct%2Fgtfs-g" -- some error
#     N/Q/R/W: "nyct%2Fgtfs-nqrw" -- works
#     1/2/3/4/5/6: "nyct%2Fgtfs" -- works
#     B/D/F/M: "nyct%2Fgtfs-bdfm" -- works
#     J/Z: "nyct%2Fgtfs-jz" -- works
#     L: "nyct%2Fgtfs-l" -- works
#     SIR (Staten Island Railway): "nyct%2Fgtfs-si"
# }
```

```python
url = base_url + "nyct%2Fgtfs-ace"
headers = {"x-api-key": API_KEY}

response = requests.get(url, headers=headers)
response.raise_for_status()

raw_data = response.content

feed = gtfs_realtime_pb2.FeedMessage()
feed.ParseFromString(raw_data)

print("Feed has", len(feed.entity), "entities.")
for entity in feed.entity[:5]:  # Just print first 5 for a quick look
    if entity.trip_update:
        print("Trip Update:", entity.trip_update.trip.trip_id)
    if entity.vehicle:
        print("Vehicle Position:", entity.vehicle.trip.trip_id)
    if entity.alert:
        print("Alert:", [t.text for t in entity.alert.header_text.translation])
```

```python
url = base_url + "nyct%2Fgtfs-si"
headers = {"x-api-key": API_KEY}

response = requests.get(url, headers=headers)
response.raise_for_status()

raw_data = response.content

feed = gtfs_realtime_pb2.FeedMessage()

print (response.headers)
print( "------------" )
print (response.content[:200])

feed.ParseFromString(raw_data)

print("Feed has", len(feed.entity), "entities.")
for entity in feed.entity[:5]:  # Just print first 5 for a quick look
    if entity.trip_update:
        trip = entity.trip_update.trip
        print(f"Trip Update:, {trip.trip_id}, Route ID: {trip.route_id}")
    if entity.vehicle:
        vehicle = entity.vehicle
        current_status = vehicle.current_status  # 0 = INCOMING_AT, 1 = STOPPED_AT, 2 = IN_TRANSIT_TO
        print(f"Vehicle Position - Trip ID: {vehicle.trip.trip_id}, Status: {current_status}")
    if entity.alert:
        alert = entity.alert
        print("Alert:", [t.text for t in alert.header_text.translation])
```
